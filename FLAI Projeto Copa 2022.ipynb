### Rank das equipes na primeira fase

```python
# RANK DE CADA GRUPO 1st
dic1 = {}
for i in (set(df['grupo'])):
    indicador = 0
    for i in (sorted(df['1st'].loc[df['grupo']==i],reverse=True)):
        indicador += 1
        dic1[i] = indicador
```

```python
# aplicar variável com valor de rank
df['rank1st'] = df['1st'].map(dic1)
```

```python
# RANK DE CADA GRUPO 2nd
dic2 = {}
for i in (set(df['grupo'])):
    indicador = 0
    for i in (sorted(df['2nd'].loc[df['grupo']==i],reverse=True)):
        indicador += 1
        dic2[i] = indicador
```

```python
# aplicar variável com valor de rank
df['rank2nd'] = df['2nd'].map(dic2)
```

```python
# variáveis
url = r'C:\Users\geanc\Downloads\DATA SCIENCE\FLAI\Programa FLAI Dominando Data Science 3.0\FLAI Projeto Copa 2022\18-11-2022\outputPlacaresPartidasPROBS.xlsx'

# jogo = ['grupoASenegalxHolanda']

jogo = [
'grupoASenegalxHolanda','grupoACatarxEquador','grupoACatarxSenegal','grupoAHolandaxEquador','grupoAHolandaxCatar','grupoAEquadorxSenegal',
'grupoBInglaterraxIrã','grupoBEstadosUnidosxPaísdeGale','grupoBPaísdeGalesxIrã','grupoBInglaterraxEstadosUnidos','grupoBPaísdeGalesxInglaterra','grupoBIrãxEstadosUnidos',
'grupoCArgentinaxArábiaSaudita','grupoCMéxicoxPolônia','grupoCPolôniaxArábiaSaudita','grupoCArgentinaxMéxico','grupoCPolôniaxArgentina','grupoCArábiaSauditaxMéxico',
'grupoDDinamarcaxTunísia','grupoDFrançaxAustrália','grupoDTunísiaxAustrália','grupoDFrançaxDinamarca','grupoDTunísiaxFrança','grupoDAustráliaxDinamarca',
'grupoEAlemanhaxJapão','grupoEEspanhaxCostaRica','grupoEJapãoxCostaRica','grupoEEspanhaxAlemanha','grupoEJapãoxEspanha','grupoECostaRicaxAlemanha',
'grupoFMarrocosxCroácia','grupoFBélgicaxCanadá','grupoFBélgicaxMarrocos','grupoFCroáciaxCanadá','grupoFCroáciaxBélgica','grupoFCanadáxMarrocos',
'grupoGSuíçaxCamarões','grupoGBrasilxSérvia','grupoGCamarõesxSérvia','grupoGBrasilxSuíça','grupoGCamarõesxBrasil','grupoGSérviaxSuíça',
'grupoHUruguaixCoreiadoSul','grupoHPortugalxGana','grupoHCoreiadoSulxGana','grupoHPortugalxUruguai','grupoHCoreiadoSulxPortugal','grupoHGanaxUruguai'
       ]

s1, s2, prob, p1, p2 = [], [], [], [], []
```

```python
for i in jogo:
    df_placar = pd.read_excel(url, sheet_name=i, skiprows=0, usecols='A:J')
    df_placar.replace('7+',7,inplace=True)

    # nome seleção 1
    se1 = df_placar.iloc[2,0]

    # nome seleção 2
    se2 = df_placar.columns[2]

    # remover linhas
    df_placar = df_placar.drop([0,1])

    # remover coluna
    df_placar.drop(columns = ['Unnamed: 0'], inplace=True)

    # renomear colunas
    df_placar.columns = ['x',0,1,2,3,4,5,6,7]

    df_placar.set_index('x',inplace=True)

    # deixar em porcentagem os valores de cada placar
    df_placar = df_placar*100

    # separar gols de cada seleção para a maior probabilidade
    # prob = np.max(df_placar[1])
    # prob = df_placar.max()[0]    
    # prob = sorted(np.max(df_placar),reverse=True)[0]
    probabilidade = sorted(df_placar.max(axis=0),reverse=True)[0]    
    pl1 = df_placar.idxmax().head(1)[0]
    pl2 = df_placar.T.idxmax().head(1)[0]

    s1.append(se1)
    s2.append(se2)
    prob.append(probabilidade)
    p1.append(pl1)
    p2.append(pl2)

placares = pd.DataFrame(list(zip(jogo,s1,p1,s2,p2,prob)), columns=['jogo','s1', 'p1', 's2','p2',' prob'])
```
